```python
class Song:
    def __init__(self, titol, autor, durada):
        self.titol = titol
        self.autor = autor
        self.durada = durada

    def info(self):
        return f"'{self.titol}' de {self.autor} ({self.durada}s)"

class Playlist:
    def __init__(self, nombre):
        self.nombre = nombre
        self.song_list = []
        
    def afegir_canco(self, song):
        self.song_list.append(song)

    def durada_total(self):
        return sum(song.durada for song in self.song_list)

    def mostrar_playlist(self):
        print(f"\n{self.nombre}")
        for i, song in enumerate(self.song_list, 1):
            print(f"{i}. {song.info()}")

    def canco_mes_llarga(self):
        if not self.song_list:
            return "La lista está vacía."
      
        llarga = max(self.song_list, key=lambda s: s.durada)
        return llarga.info()


musica = Playlist('Favoritos')

beat_it = Song('Beat it', 'Michael Jackson', 299)
the_last_ship = Song('The last ship', 'Sting', 231)
get_back = Song('Get back', 'The Beatles', 190)

musica.afegir_canco(beat_it)
musica.afegir_canco(the_last_ship)
musica.afegir_canco(get_back)

musica.mostrar_playlist()


print(f" Duración total: {musica.durada_total()} minutos")
print(f" Canción más larga: {musica.canco_mes_llarga()}")
```

```python
class Personatge:
    def __init__(self, nom, vida, forca, nivell):
        self.nom = nom
        self.vida = vida
        self.forca = forca
        self.nivell = nivell

    def atacar(self, enemic):
        enemic.vida -= self.forca
        if enemic.vida < 0: enemic.vida = 0
        print(f"{self.nom} ataca! {enemic.nom} es queda amb {enemic.vida} punts de vida.")
    def pujar_nivell(self):
         self.nivell += 1
         self.forca += 2
         self.vida += 5

    def esta_viu(self):
        if self.vida > 0:
            return True
        else:
            return False
    def __str__(self):
        return f"Nom: {self.nom} | Nivell: {self.nivell} | Vida: {self.vida} | Força: {self.forca}" 
    def info(self):
        return f"Nom: {self.nom} | Nivell: {self.nivell} | Vida: {self.vida} | Força: {self.forca}" 
        

p1 = Personatge("Jugador", 20, 5, 1)
p2 = Personatge("Enemic", 15, 4, 1)

while p1.esta_viu() and p2.esta_viu():
    p1.atacar(p2)
    if p2.esta_viu():
        p2.atacar(p1)


guanyador = p1 if p1.esta_viu() else p2
print(f"\nEL GUANYADOR ÉS {guanyador.nom}")
guanyador.pujar_nivell()
print(f"El {guanyador} puja de nivell")
print(guanyador.info())
```
